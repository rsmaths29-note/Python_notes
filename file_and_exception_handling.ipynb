# exception handling

```python
"""
we use a try except block for dealing with exceptions in the following way.
the try part basically...well tries the code under it. if it raises an exception
and we have an except block, it gets dealt with byt he code following the except block.

"""

# this should not return any output since we have b = 0.
a = 2
b = 0
try :
  m = a/b
except ZeroDivisionError :
  pass
else :
  print(m)
```

```python
"""
when dealing with more than one exceptions, you write two except blocks. also,
if you want to use the exception name in your code you use the 'except Exception as e'
"""

# just for zero division error
a = 2
b = 0
c = "4"

try :
  m = (a/b)
except Exception as e:
  print(e,"not allowed")
else :
  print(m)

# both typeerror and zero division error
a = 3
b = 5
c = 0
try :
  m = (a*b)/c
except ZeroDivisionError:
  print("a/0 not allowed")
except TypeError:
  print("type error")
else :
  print(m)
```

```python
"""
when you want to manually raise an exception, you use the 'raise' keyword.
"""

num = int(input("enter your number please !"))
if num < 0 :
  raise ValueError("negative numbers not allowed")
else :
  print(num)
```

# reading a file using python :

```python
# if you are not using pathlib for the purpose of reading a txt file, you do it the following way :
# we use a forward slash on a windows system in the file path here. this doesn't work with pathlib.
# here the with keyword closes the file automatically.
with open("/content/trial_file") as file_object:
  content = file_object.read()
  print(content)
```

```python
# if you are using pathlib, you do this :
from pathlib import Path
file_path = Path("/content/trial_file")
content = file_path.read_text(encoding="utf-8")
print(content)
```

# reading files line by line :

```python
# without pathlib
# with closes file automatically.
# open() points to the file path.
# it is then assigned to the variable file_object
with open("/content/trial_file.txt") as file_object:
  for line in file_object :
    print(line.rstrip())

```

```python
# when using pathlib, you use splitlines to get the lines in the content.
from pathlib import Path

path = Path("/content/trial_file.txt")
content = path.read_text()
for line in content.splitlines() :
  print(line.rstrip())
```

# Making a list of all the lines in a text file

```python
# without pathlib :
with open("/content/trial_file.txt") as file_object :
  line_list = file_object.readlines()

for line in line_list :
    print(line.rstrip())
```

```python
# with pathlib
from pathlib import Path

path = Path("/content/trial_file.txt")
content = path.read_text().splitlines()
print(content)
```

now we try all three together. once with and then without pathlib.

```python
# with pathlib, we try to print the contents of a file three times with three different ways

from pathlib import Path

path = Path("/content/trial_file.txt")
content = path.read_text().rstrip()
print(content)

content = path.read_text().splitlines()
for line in content :
  print(line.rstrip())
```

```python
# without pathlib we do :

with open("/content/trial_file.txt") as file_object :
  content = file_object.read()
  print(content)

with open("/content/trial_file.txt") as file_object :
  for line in file_object :
    print(line.rstrip())

with open("/content/trial_file.txt") as file_object :
  lines = file_object.readlines()
  for line in lines :
    print(line.rstrip())
```
