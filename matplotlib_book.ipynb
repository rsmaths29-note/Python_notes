```python
import matplotlib.pyplot as plt

# first we define the x and y values.
x_values = list(range(0,5000))
y_values = [x**3 for x in x_values]

# plt.style.use() is specified to use a builtin style.
plt.style.use('petroff10')

# figure refers to the entire figure in which several plots can be visualised.
# ax is one of those plots.
# think of this statement as first making a box for all your plots.
# and think of drawing a little graph ax in the box.
figure, ax = plt.subplots()

# we use ax.scatter to plot a scatter plot.
# we give the x values and the y values.
# s=3 stands for size = 3
# c=y_values assign the y values to be the one for which we apply the colormap.
# cmap = plt.cm.Blues of course, specifies the color map.
ax.scatter(x_values, y_values, s=3, c=y_values, cmap=plt.cm.Blues)

# ax.set_title gives the plot title.
# set_xlabel and set_ylabel specify the x and y labels for our axes.
ax.set_title("cube values", fontsize=20)
ax.set_xlabel("values", fontsize=11)
ax.set_ylabel("cubes", fontsize=11)

# ax.axis()takes a list of four numbers. this specifies the ranges of our plot
ax.axis([0, 5001, 0, 5000**3])

# ax.tick_params() is used to specify tick parameters.
# ax.ticklabel_format() is use to specify that hey, dont use scientific notation for large numbers.
ax.tick_params(labelsize=15)
ax.ticklabel_format(style="plain")

# plt.show() shows the plot in a window.
plt.show()
```

```python
from random import choice
from random import randint
import matplotlib.pyplot as plt

class RandomWalk :
  """a class to model random walks"""
  def __init__(self, num_points=5000):
    self.num_points = num_points
    self.x_values = [0]
    self.y_values = [0]

  def walk(self) :
    """activate the random walk"""
    while len(self.x_values) < self.num_points :
      x_direction = choice([-1, 1])
      x_distance = choice([0,1,2,3,4,5])
      x_total_distance = x_direction*x_distance

      y_direction = choice([-1, 1])
      y_distance = choice([0,1,2,3,4,5])
      y_total_distance = y_direction*y_distance

      if x_total_distance == 0 and y_total_distance == 0 :
        continue

      x = self.x_values[-1] + x_total_distance
      y = self.y_values[-1] + y_total_distance

      self.x_values.append(x)
      self.y_values.append(y)

rw = RandomWalk()
rw.walk()

fig, ax = plt.subplots()
ax.plot(rw.x_values, rw.y_values, linewidth=3)
ax.set_title("Randomwalk", fontsize=21)
ax.set_xlabel("x", fontsize=11)
ax.set_ylabel("y", fontsize = 11)
ax.set_aspect("equal")
ax.tick_params(labelsize=15)
ax.ticklabel_format(style="plain")

plt.show()
```

```python
import matplotlib.pyplot as plt
from random import choice

class Coin:
  """a class to model a coin"""
  def __init__(self, head_weight=1, tail_weight=1) :
    """initialise its attributes"""
    self.head_weight = head_weight
    self.tail_weight = tail_weight
    self.num_heads = 0
    self.num_tails = 0

  def flip(self) :
    outcome = choice([-1, 1])
    return outcome

  def flip_sequence(self, points=50000) :
    outcome_list = [0]
    for x in range(0, points) :
      y = self.flip()
      if y == 1 :
        self.num_heads += 1
      else :
        self.num_tails += 1
      posiion = outcome_list[-1] + y
      outcome_list.append(posiion)
    return outcome_list

x_values = range(0, 50001)
my_coin = Coin()
experiment = my_coin.flip_sequence()

fig, ax = plt.subplots()
plt.figure(figsize=(20, 4))
ax.scatter(x_values, experiment, s=1, color=(1, 0.4, 0.8))
ax.set_title("flips", fontsize=22)
ax.set_xlabel("step", fontsize=11)
ax.set_ylabel("flips", fontsize=11)
ax.tick_params(labelsize=10)
ax.ticklabel_format(style="plain")


plt.show

print("number of heads : ", my_coin.num_heads)
print("number of tails : ", my_coin.num_tails)
print("number of flips :  50000")
```

## styling the random walk

```python
from random import choice
import matplotlib.pyplot as plt

def move(max_distance=5) :
  """direction and distance covered across an axis"""
  direction = choice([-1,1])
  distance = choice(list(range(0, max_distance+1)))
  vector = direction*distance
  return vector


class RandomWalk :
  """Class to model a random walk on the x-y plane"""

  def __init__(self, num_points=5000) :
    """initialise the attributes of a random walk"""
    # num_points represent the no. of times we hop onto a new point.
    self.num_points = num_points
    # the following lists will store the coordinates of every point in our random walk.
    # all walks start at (0,0)
    self.x_coordinates = [0]
    self.y_coordinates = [0]

  def walk(self) :
    """walk at a given point"""
    x_move = move()
    y_move = move()
    return (x_move, y_move)


rw = RandomWalk()

while len(rw.x_coordinates) < rw.num_points :
  step = rw.walk()
  if step == (0,0) :
    continue
  else :
    x_position = rw.x_coordinates[-1] + step[0]
    y_position = rw.y_coordinates[-1] + step[1]
    rw.x_coordinates.append(x_position)
    rw.y_coordinates.append(y_position)

point_no = range(rw.num_points)
fig, ax = plt.subplots()
ax.scatter(rw.x_coordinates, rw.y_coordinates, c=point_no, cmap=plt.cm.Blues, edgecolors="none", s=15)
ax.set_aspect("equal")

plt.show()



```
